Узнаем, насколько лояльны пользователи к нашему сервису и какой их процент пользуется им многократно. Для этого расчитаем Retention Rate, разделив пользователей на месячные когорты. После выполнения всех вычислений построим тепловую карту для более наглядного отображения Retention Rate по когортам.

```python
visits['activity_month']=visits['start_ts'].astype('datetime64[M]')
visits.head()
```

```python
first_activity_month=visits.groupby('uid')['activity_month'].min()
first_activity_month.name='first_activity_month'
visits=visits.merge(first_activity_month, on ='uid')
visits.head()
```

```python
visits['cohort_lifetime']=visits['activity_month']-visits['first_activity_month']
visits['cohort_lifetime']=visits['cohort_lifetime']/np.timedelta64(1, 'M')
visits['cohort_lifetime']=visits['cohort_lifetime'].astype('float').round(0)
```

```python
visits.head()
```

```python
cohorts=visits.groupby(['first_activity_month', 'cohort_lifetime'])['uid'].nunique()
initial_users_count = cohorts.reset_index().query('cohort_lifetime==0')[['first_activity_month', 'uid']]
initial_users_count=initial_users_count.rename(columns={'uid':'cohort_users'})
cohorts=cohorts.reset_index().merge(initial_users_count, on='first_activity_month')
cohorts['retention']=cohorts['uid']/cohorts['cohort_users']
cohorts['retention']=cohorts['retention']
cohorts.head()
```

Для построения тепловой карты создадим сводную таблицу cohorts_ret_rate, где индексами будут месяцы первой активности когорт, столбцами - срок жизни когорт, значениями - данные из столбца `retention`, а аггрегирующей функцией будет выступать среднее значение этого показателя.

```python
cohorts['cohort_lifetime']=cohorts['cohort_lifetime'].astype('int')
cohorts_ret_rate=cohorts.pivot_table(index='first_activity_month',
                                    columns='cohort_lifetime',
                                    values='retention',
                                    aggfunc='sum')
```

```python
import seaborn as sns
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('User Retention by Cohorts')
sns.heatmap(cohorts_ret_rate, annot=True, fmt='.1%', linewidths=1, linecolor='gray', vmax=0.1)
```

Как видим, Retention Rate Яндекс.Афиш не очень высок. Уже на второй месяц у всех когорт он падает до уровня меньше 10%, далее не поднимаяясь выше 7%. ## ПРОДАЖИ

Выясним, сколько времени требуется среднему пользователю для совершения покупки на сервисе. Для этого объединим таблицы visits и orders методом merge и посчитаем разницу между временем покупки (столбец `buy_ts`) и времнем захода на сервис (столбец `start_ts`) в минутах.

```python
users=visits.sort_values(by='start_ts').groupby('uid').agg({'start_ts':'first', 'source_id':'first', 'device':'first'})
orders1=orders.sort_values(by='buy_ts').groupby('uid').agg({'buy_ts':'first', 'revenue':'sum'})
sales=users.merge(orders1, on='uid')
```

## Расчет LTV, CAC И ROMI по источникам привлечения пользователей

В этой части исследования мы постоим тепловые для показателей LTV и ROMI по когортам, а также график исменения САС для каждого маркетингового источника по месяцам. Для этого используем созданные раннее таблицы visits_new и source_cac.

Посмотрим с помощью следующего цикла тепловые карты LTV для каждого маркетингового источника, объединив покупателей в месячные когорты.

```python
#Построение графиков LTV
visits_new=visits_new.sort_values('source_id')
visits_new['order_month'] = visits_new['buy_ts'].astype('datetime64[M]')
for i in visits_new['source_id'].unique():
    if i!=7:
        visits_new_i=visits_new.loc[visits_new['source_id']==i]
        first_order_month_i=visits_new_i.groupby('uid')['order_month'].min().reset_index()
        first_order_month_i.columns=['uid', 'first_order_month']
        visits_new_i_ltv=visits_new_i.merge(first_order_month, on='uid')
        cohort_sizes_i=first_order_month_i.groupby('first_order_month')['uid'].nunique().reset_index()
        cohort_sizes_i.columns=['first_order_month', 'total_buyers']
        cohorts_i = visits_new_i_ltv.groupby(['first_order_month','order_month'])['revenue'].sum().reset_index()
        report_i=cohorts_i.merge(cohort_sizes_i, on='first_order_month')
        report_i['age'] = (report_i['order_month'] - report_i['first_order_month']) / np.timedelta64(1, 'M')
        report_i['age'] = report_i['age'].round().astype('int')
        report_i['ltv']=report_i['revenue']/report_i['total_buyers']
        output_i=report_i.pivot_table(index='first_order_month',
                         columns='age',
                         values='ltv',
                         aggfunc='mean').round(2)
        sns.set(style='white')
        plt.figure(figsize=(16, 9))
        plt.title('LTV по источнику {}'.format(i))
        sns.heatmap(output_i.cumsum(axis=1).round(2), annot=True, fmt=".2f", linewidths=1, linecolor='gray')     
```

Несмотря на то, что на источник 3 тратится больше всего средств, показатель LTV по этому источнику уступает практически всем остальным. Среди всех источников по показателю LTV особенно можно выделить источник 1. Проверим рентабельность этих и других источников на следующих тепловых картах.

```python
#Построение графиков ROMI
for i in visits_new['source_id'].unique():
    if i!=7:
        visits_new_i=visits_new.loc[visits_new['source_id']==i]
        first_order_month_i=visits_new_i.groupby('uid')['order_month'].min().reset_index()
        first_order_month_i.columns=['uid', 'first_order_month']
        visits_new_i_ltv=visits_new_i.merge(first_order_month, on='uid')
        cohort_sizes_i=first_order_month_i.groupby('first_order_month')['uid'].nunique().reset_index()
        cohort_sizes_i.columns=['first_order_month', 'total_buyers']
        cohorts_i = visits_new_i_ltv.groupby(['first_order_month','order_month'])['revenue'].sum().reset_index()
        report_i=cohorts_i.merge(cohort_sizes_i, on='first_order_month')
        report_i['age'] = (report_i['order_month'] - report_i['first_order_month']) / np.timedelta64(1, 'M')
        report_i['age'] = report_i['age'].round().astype('int')
        report_i['ltv']=report_i['revenue']/report_i['total_buyers']
        source_cac_i=source_cac[source_cac['source_id']==i].reset_index()['cac'][0]
        report_i['romi']=report_i['ltv']/source_cac_i
        output_i_romi=report_i.pivot_table(index='first_order_month',
                         columns='age',
                         values='romi',
                         aggfunc='mean').round(2)
        sns.set(style='white')
        plt.figure(figsize=(16, 9))
        plt.title('ROMI по источнику {}'.format(i))
        sns.heatmap(output_i_romi.cumsum(axis=1).round(2), annot=True, fmt=".2f", linewidths=1, linecolor='gray', vmin=0.5)
```

Тепловые карты демонстрируют потрясающие показатели ROMI для источника 1 и совсем неубедительные для источника 3. Показатель ROMI превышает единицу для первых четырех когорт покупателей у источников 4, 5 и 9, притом что на источник 9, как мы могли убедиться выше, выделяется совсем мало средств.
